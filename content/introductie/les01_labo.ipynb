## NumPy en Matplotlib

**NumPy** (Numerical Python) is dé library voor numerieke berekeningen in Python. Het biedt krachtige datastructuren zoals arrays en matrices, en functies voor wiskundige operaties. In machine learning wordt NumPy gebruikt om data voor te stellen en te manipuleren. Bijna alle ML-libraries (PyTorch, TensorFlow, scikit-learn) zijn gebouwd op NumPy.

**Matplotlib** is de standaard library voor visualisatie in Python. Hiermee maak je grafieken, plots en afbeeldingen. Visualisatie is essentieel om data te begrijpen en resultaten te communiceren.
