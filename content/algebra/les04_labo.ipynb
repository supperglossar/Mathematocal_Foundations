## Oefening 1: Stelsels Opstellen


### Opdracht 1a

Vertaal de volgende woordproblemen naar matrixvorm Ax = b:

**Probleem 1:** Een winkel verkoopt appels en peren. 3 appels en 2 peren kosten €5. 1 appel en 4 peren kosten €6. Wat is de prijs per appel en per peer?

**Probleem 2:** Een trein rijdt van A naar B. Met de wind mee (snelheid + wind) doet hij er 2 uur over voor 300 km. Tegen de wind in (snelheid - wind) doet hij er 3 uur over voor dezelfde afstand. Wat is de snelheid van de trein en de windsnelheid?

````python
# Jouw code voor opdracht 1a:

# Probleem 1: appels (a) en peren (p)
# 3a + 2p = 5
# 1a + 4p = 6


# Probleem 2: snelheid (s) en wind (w)
# (s + w) * 2 = 300  →  2s + 2w = 300
# (s - w) * 3 = 300  →  3s - 3w = 300

A = np.array([[3, 2],
              [1, 4]])
b = np.array([5, 6])

print("Stelsel van probleem 1:")
print("  3a + 2p = 5")
print("  1a + 4p = 6")
print()
print("In matrixvorm Ax = b:")
print(f"A = \n{A}")
print(f"b = {b}")
print()
print("We zoeken x = [x, y] zodat A @ x = b")

S = np.array([[2, 2],
              [3, 3]])
w = np.array([300, 300])

print("Stelsel van probleem 2:")
print("  2s + 2w = 300")
print("  3s - 3w = 300")
print()
print("In matrixvorm Ax = b:")
print(f"A = \n{S}")
print(f"b = {w}")
print()
print("We zoeken x = [x, y] zodat A @ x = b")
````

### Opdracht 1b

Los beide stelsels op met `np.linalg.solve()` en interpreteer de antwoorden.

````python
# Jouw code voor opdracht 1b:

# Oplossen met NumPy
x_oplossing = np.linalg.solve(A, b)

print(f"Oplossing 1: x = {x_oplossing}")
print(f"  x = {x_oplossing[0]}")
print(f"  y = {x_oplossing[1]}")
print()

# Verificatie
print("Verificatie (A @ x moet gelijk zijn aan b):")
print(f"  A @ x = {A @ x_oplossing}")
print(f"  b     = {b}")
print(f"  Gelijk? {np.allclose(A @ x_oplossing, b)}")

````

### Opdracht 2a

Los de volgende stelsels op en verifieer je antwoord door A @ x te berekenen:

**Stelsel 1:**
```
2x + y = 5

```

**Stelsel 2:**

````python
# Jouw code voor opdracht 2a:

# Stelsel 1:
# 2x + y = 5
# x - y = 1
A1 = np.array([[2, 1],
               [1, -1]])
b1 = np.array([5, 1])

x1_solution = np.linalg.solve(A1, b1)

print(f"Oplossing voor Stelsel 1: x = {x1_solution}")
print(f"  x = {x1_solution[0]:.2f}")
print(f"  y = {x1_solution[1]:.2f}")

# Verificatie
print("Verificatie (A @ x moet gelijk zijn aan b):")
print(f"  A @ x = {A1 @ x1_solution}")
print(f"  b     = {b1}")
print(f"  Gelijk? {np.allclose(A1 @ x1_solution, b1)}")

# Stelsel 2:
# x + y + z = 6
# 2x - y + z = 3
# x + 2y - z = 3
A2 = np.array([[1, 1, 1],
               [2, -1, 1],
               [1, 2, -1]])
b2 = np.array([6, 3, 3])

x2_solution = np.linalg.solve(A2, b2)

print(f"\nOplossing voor Stelsel 2: x = {x2_solution}")
print(f"  x = {x2_solution[0]:.2f}")
print(f"  y = {x2_solution[1]:.2f}")
print(f"  z = {x2_solution[2]:.2f}")

# Verificatie
print("Verificatie (A @ x moet gelijk zijn aan b):")
print(f"  A @ x = {A2 @ x2_solution}")
print(f"  b     = {b2}")
print(f"  Gelijk? {np.allclose(A2 @ x2_solution, b2)}")
````

### Opdracht 2b

Wat gebeurt er als je probeert het volgende stelsel op te lossen?

```
x + 2y = 3
2x + 4y = 6
```

Leg uit waarom dit stelsel oneindig veel oplossingen heeft.

````python
# Jouw code voor opdracht 2b:

A_inf = np.array([[1, 2],
                  [2, 4]])
b_inf = np.array([3, 6])

print("Matrix A:")
print(A_inf)
print("Vector b:")
print(b_inf)

determinant_inf = np.linalg.det(A_inf)
print(f"De determinant van A is: {determinant_inf}")

print("\nDit stelsel heeft oneindig veel oplossingen omdat de tweede vergelijking een veelvoud is van de eerste vergelijking.")
print("2 * (x + 2y) = 2 * 3  wordt  2x + 4y = 6.")
print("Beide vergelijkingen stellen dezelfde lijn voor in een 2D-vlak, dus elk punt op die lijn is een oplossing.")
print("Een determinant van 0 bevestigt dat de matrix singulier is en dus niet uniek oplosbaar (of geen oplossing, of oneindig veel).")

try:
    x_inf = np.linalg.solve(A_inf, b_inf)
    print(f"Oplossing: {x_inf}")
except np.linalg.LinAlgError as e:
    print(f"\nFout bij het oplossen van het stelsel met np.linalg.solve(): {e}")
    print("Dit bevestigt dat de matrix singulier is en niet uniek oplosbaar.")
````

### Opdracht 2c

Wat gebeurt er als je probeert het volgende stelsel op te lossen?

```
x + 2y = 3
2x + 4y = 8
```

Leg uit waarom dit stelsel geen oplossing heeft.

````python
# Jouw code voor opdracht 2c:

A_no_sol = np.array([[1, 2],
                     [2, 4]])
b_no_sol = np.array([3, 8])

print("Matrix A:")
print(A_no_sol)
print("Vector b:")
print(b_no_sol)

determinant_no_sol = np.linalg.det(A_no_sol)
print(f"De determinant van A is: {determinant_no_sol}")

print("\nDit stelsel heeft geen oplossing. Hoewel de linkerzijde van de tweede vergelijking een veelvoud is van de eerste (2x + 4y = 2 * (x + 2y)), is de rechterzijde dat niet (8 != 2 * 3).")
print("Dit betekent dat de twee vergelijkingen parallelle lijnen beschrijven die nooit snijden, waardoor er geen gemeenschappelijk punt is dat aan beide vergelijkingen voldoet.")
print("Een determinant van 0 bevestigt dat de matrix singulier is, wat kan leiden tot geen oplossing of oneindig veel oplossingen.")

try:
    x_no_sol = np.linalg.solve(A_no_sol, b_no_sol)
    print(f"Oplossing: {x_no_sol}")
except np.linalg.LinAlgError as e:
    print(f"\nFout bij het oplossen van het stelsel met np.linalg.solve(): {e}")
    print("Dit bevestigt dat er geen unieke oplossing is, en in dit specifieke geval geen enkele oplossing.")
````

## Oefening 3: De Inverse Berekenen


### Opdracht 3a

Bereken de inverse van de volgende 2×2 matrix met de hand (gebruik de formule) en verifieer met NumPy:

```
A = [[4, 7],
     [2, 6]]
```

Formule: A⁻¹ = (1/det(A)) × [[d, -b], [-c, a]] voor A = [[a, b], [c, d]]

````python
# Jouw code voor opdracht 3a:

# 2×2 inverse met de hand berekenen
A = np.array([[4, 7],
              [2, 6]])

a, b = A[0, 0], A[0, 1]
c, d = A[1, 0], A[1, 1]

# Determinant
det = a*d - b*c

# Inverse volgens formule
A_inv_manual = (1 / det) * np.array([[d, -b],
                                      [-c, a]])

print(f"Matrix A:")
print(A)
print()
print(f"Determinant: ad - bc = {a}×{d} - {b}×{c} = {det}")
print()
print("Inverse (handmatig):")
print(A_inv_manual)
print()
print("Inverse (NumPy):")
print(np.linalg.inv(A))
print()
print(f"Gelijk? {np.allclose(A_inv_manual, np.linalg.inv(A))}")
````

### Opdracht 3b

Verifieer dat A @ A⁻¹ = I en A⁻¹ @ A = I voor de matrix uit 3a.

````python
# Jouw code voor opdracht 3b:

# Gebruik de eerder berekende A en A_inv_manual
A = np.array([[4, 7],
              [2, 6]])
det = np.linalg.det(A)
A_inv_manual = (1 / det) * np.array([[6, -7],
                                      [-2, 4]])

identity_matrix = np.identity(2)

print("Verificatie A @ A⁻¹ = I:")
product1 = A @ A_inv_manual
print(f"A @ A⁻¹ = \n{product1}")
print(f"Is gelijk aan identiteitsmatrix? {np.allclose(product1, identity_matrix)}")

print("\nVerificatie A⁻¹ @ A = I:")
product2 = A_inv_manual @ A
print(f"A⁻¹ @ A = \n{product2}")
print(f"Is gelijk aan identiteitsmatrix? {np.allclose(product2, identity_matrix)}")
````

### Opdracht 3d

Gebruik de inverse om het stelsel Ax = b op te lossen via x = A⁻¹b:

```
A = [[4, 7],
     [2, 6]]
b = [5, 3]
```

Vergelijk met de oplossing via `np.linalg.solve()`.

````python
# Jouw code voor opdracht 3d:

A = np.array([[4, 7],
              [2, 6]])
b = np.array([5, 3])

print(f"Matrix A:\n{A}")
print(f"Vector b: {b}")

# Oplossing via de inverse matrix: x = A⁻¹b
A_inv = np.linalg.inv(A)
x_inv = A_inv @ b
print(f"\nOplossing via inverse matrix (x = A⁻¹b): {x_inv}")

# Oplossing via np.linalg.solve()
x_solve = np.linalg.solve(A, b)
print(f"Oplossing via np.linalg.solve():     {x_solve}")

# Vergelijking van de resultaten
print(f"\nZijn de oplossingen gelijk? {np.allclose(x_inv, x_solve)}")
````
