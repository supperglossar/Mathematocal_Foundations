## Oefening 1: Matrices Aanmaken


### Opdracht 1a

Maak de volgende matrices aan met NumPy functies:

1. Een 3×4 matrix gevuld met nullen
2. Een 2×5 matrix gevuld met enen
3. Een 4×4 identiteitsmatrix
4. Een 3×3 diagonaalmatrix met waarden [2, 5, 7] op de diagonaal
5. Een 3×4 matrix met random waarden uit een normale verdeling

````python
# Jouw code voor opdracht 1a:

# 1. Een 3x4 matrix gevuld met nullen
matrix_zeros = np.zeros((3, 4))
print("3x4 matrix met nullen:")
print(matrix_zeros)
print()

# 2. Een 2x5 matrix gevuld met enen
matrix_ones = np.ones((2, 5))
print("2x5 matrix met enen:")
print(matrix_ones)
print()

# 3. Een 4x4 identiteitsmatrix
identity_matrix = np.eye(4)
print("4x4 identiteitsmatrix:")
print(identity_matrix)
print()

# 4. Een 3x3 diagonaalmatrix met waarden [2, 5, 7] op de diagonaal
diagonal_matrix = np.diag([2, 5, 7])
print("3x3 diagonaalmatrix met [2, 5, 7]:")
print(diagonal_matrix)
print()

# 5. Een 3x4 matrix met random waarden uit een normale verdeling
random_matrix = np.random.randn(3, 4)
print("3x4 matrix met random waarden (normale verdeling):")
print(random_matrix)
print()
````

### Opdracht 1b

Gegeven de vector `v = np.arange(12)` (getallen 0-11), reshape deze naar:

1. Een 3×4 matrix
2. Een 4×3 matrix
3. Een 2×6 matrix
4. Een 2×2×3 tensor (3D)

Print elke keer de shape en de matrix zelf.

````python
v = np.arange(12)
print(f"Originele vector: {v}\n")

# Jouw code voor opdracht 1b:

# Matrix een
matrix_een = v.reshape(3, 4)
print("3x4 matrix:")
print(matrix_een)
print()

# Matrix twee
matrix_twee = v.reshape(4, 3)
print("4x3 matrix:")
print(matrix_twee)
print()

# Matrix drie
matrix_drie = v.reshape(2, 6)
print("2x6 matrix:")
print(matrix_drie)
print

# Matrix vier
matrix_vier = v.reshape(2, 2, 3)
print("2x2x3 tensor:")
print(matrix_vier)
````
